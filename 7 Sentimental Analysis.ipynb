```python
for text in texts:
    sentiment = TextBlob(text).sentiment
    if sentiment.polarity > 0:
        print(f"Text: {text}\nSentiment: Positive\n")
    elif sentiment.polarity < 0:
        print(f"Text: {text}\nSentiment: Negative\n")
    else:
        print(f"Text: {text}\nSentiment: Neutral\n")
    print()
```
