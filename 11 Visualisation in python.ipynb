```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Matplotlib
* Line, Bar, Histogram, Scatter, Pie, Box, Voilin, Heatmap

### Line Plot

```python
x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave')
plt.show()
```

### Bar Plot

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

### Histogram

```python
data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

### Scatter Plot

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

### Pie Plot

```python
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 25, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

### Box Plot

```python
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 1, 100)
data3 = np.random.normal(2, 1, 100)

plt.boxplot([data1, data2, data3], labels=['Group 1', 'Group 2', 'Group 3'])
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Box Plot with Multiple Boxes')
plt.show()
```

### Violin Plot

```python
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 1, 100)
data3 = np.random.normal(2, 1, 100)

plt.violinplot([data1, data2, data3], showmeans=False, showmedians=True)
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3'])
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Violin Plot using Matplotlib')
plt.grid(axis='y')
plt.show()
```

### Heatmap 

```python
data = np.random.rand(10, 10)

plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap using Matplotlib')
plt.show()
```

### Line Plot 

```python
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Line plot
sns.lineplot(data=data, x='X', y='Y')
plt.title('Line Plot with Seaborn')
plt.show()
```

### Joint Plot 

```python
np.random.seed(0)
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Joint plot
sns.jointplot(data=data, x='X', y='Y', kind='scatter')
plt.suptitle('Joint Plot with Seaborn')
plt.show()
```

### Bar Plot 

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Create a DataFrame
data = pd.DataFrame({'Categories': categories, 'Values': values})

# Bar plot
sns.barplot(data=data, x='Categories', y='Values')
plt.title('Bar Plot with Seaborn')
plt.show()
```

### Box Plot

```python
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 1, 100)
data3 = np.random.normal(2, 1, 100)

data = {'Group 1': data1, 'Group 2': data2, 'Group 3': data3}
df = pd.DataFrame(data)

sns.boxplot(data=df)
plt.title('Box Plot with Multiple Boxes (Seaborn)')
plt.show()
```

### Violin Plot

```python
data = np.random.normal(0, 1, (100, 3))
labels = ['A', 'B', 'C']

sns.violinplot(data=data)
plt.xticks(ticks=[0, 1, 2], labels=labels)
plt.title('Violin Plot')
plt.show()
```

### Heatmap

```python
data = np.random.rand(10, 10)

sns.heatmap(data, annot=True)
plt.title('Heatmap')
plt.show()
```

### Time Series

```python
# Generate sample non-linear time series data
np.random.seed(0)
dates = pd.date_range('2022-01-01', periods=100)
values = np.sin(np.linspace(0, 10, 100)) + np.random.normal(0, 0.1, 100)
values = pd.Series(values, index=dates)

# Create a DataFrame
data = pd.DataFrame({'Date': dates, 'Value': values})

# Line plot for non-linear time series
sns.lineplot(data=data, x='Date', y='Value')
plt.title('Non-Linear Time Series Plot with Seaborn')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```
